Import the necessary modules and packages

```python
import os
```

```python

from textflow.SequenceDirectory import SequenceDirectory
from textflow.SequenceFile import SequenceFile
from textflow.SequenceString import SequenceString
from textflow.POSAnalyzer import POSAnalyzer
```

Create a Sequence from a directory:

```python
s = SequenceDirectory(os.getcwd()+"/ExampleDirectory",["directories","files","words"],[SequenceFile,SequenceString])
```

```python
s
```

Create a POS analyzer and apply it to a sequence:
+ Store the result of the analyzer in the same level as the analysis level 

```python
posAnalyzer = POSAnalyzer()
posAnalyzer.analyze(s,"POS","files/text")
s
```

+ Store the result of the analyzer in higher levels than the analysis level

```python
s = SequenceDirectory(os.getcwd()+"/ExampleDirectory")
posAnalyzer = POSAnalyzer()
posAnalyzer.analyze(s,"POS","text","files")
s
```

Use of the filter method of a sequence:

  Criteria function

```python
s = SequenceDirectory(os.getcwd()+"/ExampleDirectory")
```

```python
def sequenciasMenores4letras(arraySequencias):
  result=[]
  for i in arraySequencias:
    if len(i.metadata["text"]) < 4:
      result.append(i)
  return result
```

Using de filter

```python
for i in s.filter("files/files",sequenciasMenores4letras):
  print(i)
  print(type(i))
```

Creating a new criteria function for use in filtermetada function:

```python
def sequenciasMenores4letrasMetadata(arraySequencias):
  result=[]
  for i in arraySequencias:
    if len(i) < 4:
      result.append(i)
  return result
```

```python
for i in s.filterMetadata("files/files/text", sequenciasMenores4letrasMetadata):
  print(i)
  print(type(i))
```
