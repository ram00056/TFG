# Example 1: Analyze a Sequence with all the analyzers

```python
import pandas as pd

nombreArchivo = 'transcripts.pd'
df = pd.read_pickle(nombreArchivo)
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
nltk.download('stopwords')

listSequence = []
for index, row in df.iterrows(): 
  s = SequenceString(row["text"])
  listSequence.append(s)
  
#Creamos los analizadores
v = VolumetryAnalyzer()
st = StylometryAnalyzer(stopwords.words('spanish'),tokenizer = WordPunctTokenizer() )
pos = POSAnalyzer()
p = PolarityAnalyzer()
l = LemmaAnalyzer()
e = EmotionAnalyzer()
c = ComplexityAnalyzer()
ner = NERAnalyzer()
ir = IronityAnalyzer()
emoji = EmojiAnalyzer()
n = NGramsAnalyzer()
```

```python
for s in listSequence:
  v.analyze(s,"volumetry","text")
  st.analyze(s,"stylometry","text")
  pos.analyze(s,"POS","text")
  p.analyze(s,"polarity","text")
  l.analyze(s,"lemmas","text")
  e.analyze(s,"emotions","text")
  c.analyze(s,"complexity","text")
  ner.analyze(s,"NER","text")
  ir.analyze(s,"ironity","text")
  emoji.analyze(s,"emoji","text")
  n.analyze(s,"n-grams","text")
```

```python
s
```

```python

labels = ["volumetry","stylometry","POS","polarity","lemmas","emotions","complexity","NER","ironity","emoji","n-grams"]

dicTotal = {}

for seq in listSequence:
  for l in seq.metadata:
    if l in labels:
      if l in ["polarity", "emotions", "ironity"]:
        for category in seq.metadata[l][0]:
          l_key = str(l+category['label'])
          dicTotal[l_key] = dicTotal.get(l_key, []) + [category['score']]
      else:
        for key in seq.metadata[l][0]:
          l_key = str(l+key)
          dicTotal[l_key] = dicTotal.get(l_key, []) + [seq.metadata[l][0][key]]

for k in dicTotal:
  df[str(k)] = dicTotal[k]

df
```

```python
df.to_csv('transcript.csv')
df.to_pickle('transcriptsCompleto.pd')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
nltk.download('stopwords')

listSequence = []
for index, row in df.iterrows(): 
  s = SequenceString(row["text"])
  listSequence.append(s)
#Creamos los analizadores
v = VolumetryAnalyzer()
st = StylometryAnalyzer(stopwords.words('spanish'),tokenizer = WordPunctTokenizer() )
pos = POSAnalyzer()
p = PolarityAnalyzer(allScores=False)
l = LemmaAnalyzer()
e = EmotionAnalyzer(allScores=False)
c = ComplexityAnalyzer()
ner = NERAnalyzer()
ir = IronityAnalyzer(allScores=False)
emoji = EmojiAnalyzer()
n = NGramsAnalyzer()
```

# Construir una Sequencia Asociada al Dataframe:

En este caso, elaboramos una sequencia cuyos hijos son los distintos ids que tenemos y cada id tiene asociado las preguntas que este contesta y sus etiquetas.

```python
from textflow.SequenceToken import SequenceToken

s = SequenceToken([],[],"")
del s.metadata['text']


listSequence = []
for index, row in df.iterrows():
  if index in s.children:
    subSeq = SequenceString(row["text"],[],[])
    subSeq.metadata["question"]= row["question"]
    subSeq.metadata["state"]= row["state"]
    subSeq.metadata["label"]= row["label"]
    s.children[index].append(subSeq)
  else:
    subSeq = SequenceString(row["text"],[],[])
    subSeq.metadata["question"]= row["question"]
    subSeq.metadata["state"]= row["state"]
    subSeq.metadata["label"]= row["label"]
    s.children[index] = [subSeq]

s
```

```python
#Creamos los analizadores
v = VolumetryAnalyzer()
st = StylometryAnalyzer(stopwords.words('spanish'),tokenizer = WordPunctTokenizer() )
pos = POSAnalyzer()
p = PolarityAnalyzer()
l = LemmaAnalyzer()
e = EmotionAnalyzer()
c = ComplexityAnalyzer()
ner = NERAnalyzer()
ir = IronityAnalyzer()
emoji = EmojiAnalyzer()
n = NGramsAnalyzer()
```

```python
#Aplicamos el analizador a cada children de la sequencia
for se in s.children:
  for seq in s.children[se]:
    v.analyze(seq,"volumetry","text")
    st.analyze(seq,"stylometry","text")
    pos.analyze(seq,"POS","text")
    p.analyze(seq,"polarity","text")
    l.analyze(seq,"lemmas","text")
    e.analyze(seq,"emotions","text")
    c.analyze(seq,"complexity","text")
    ner.analyze(seq,"NER","text")
    ir.analyze(seq,"ironity","text")
    emoji.analyze(seq,"emoji","text")
    n.analyze(seq,"n-grams","text")

s
```
